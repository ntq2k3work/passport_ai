## 1. Literature Review

For video survellience tasks, head detections is one of the most important tasks. It can be used to develop tracking systems, crowd counting and crowd density estimation [1]. However, detecting heads with full variation of camera view points, poses, appearances, and scales is still a challenging task to do [2]. In the recent decades, convolutional neural networks (CNN) have gained achieved a tremendous success for object detection and classification tasks [2]. 

In this project, we built a Yolov8 based head detection model that aims to detect human heads in a crowd. The model was trained and evaluated using a dataset that can be found on Kaggle. The dataset includes morethan **14000** training images, and morethan **3000** validation images. The dataset includes annotations of two classes (head and person); however, for the sake of this project, only the head class is utilized. 

In this report, we will go over the influence of changing the hyperparameters to maximize the performance of our head detection model. 

## 2. Results and Analysis

### *Selecting an Optimizer*

The first test we did was selecting which optimizer is the best for our model. YoloV8 allows us to select from the following set of optimizers: SGD, Adam, Adamax, AdamW, NAdam, RAdam, RMSProp, or auto. When selecting auto, the Ultralytics library either selects AdamW, or SGD based on the dataset. For our case it selected SGD, hence its results will be associated with SGD. Keeping all other hyperparameters the same, the results showed that SGD and AdamW showed the best performance as compared to others. **SGD** had a slightly better results as shown below. Meanwhile, RMSProp performed worst. The details are as shown below.

```python
import pandas as pd
import matplotlib.pyplot as plt
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(y[i], i-0.1, y[i])
data = {"Models": ["RMSProp", "Adam", "NAdam", "RAdam", "AdamW", "Adamax", "SGD"],
        "mAP @0.5": [0.472, 0.671, 0.671, 0.673, 0.685, 0.686, 0.688]
       }
df = pd.DataFrame(data)
```

```python
df.plot.barh(x='Models', y='mAP @0.5', color='orange')
plt.title('Performance of Models after 20 epochs')
plt.xlabel('Model');plt.ylabel('mAP @0.5');plt.xlim([0, 0.76]);addlabels(df['Models'],df['mAP @0.5'])
plt.show()
```

The results showed that most most of them fell in the range of 0.6 - 0.7 except for the model that was optimized using RMSProp. **SGD** did the best as compared to all the other models. Hence, we are going to select SGD as our optimizer for the rest of the experiments.

### *Effect of changing batch size*

We tried to select the optimum batch size that can optimize our VRAM usage. However, the results showed that a batch size of 64 worked best with our training data. The expermiment was done using the YoloV8 nano model size, and batch sizes ranging from **4** to **128**. The results are as shown below.

```python
import pandas as pd
import matplotlib.pyplot as plt
# mAP @0.5 for SGD as we increase the number of epochs
data = {"Batch Size": [2, 8, 32, 64, 128],
        "mAP @0.5": [0.667, 0.685, 0.687, 0.687, 0.624]
       }
df = pd.DataFrame(data)
```

```python
plt.plot(df['Batch Size'], df['mAP @0.5'], color='purple')
plt.title('Performance of Models after 20 epochs');plt.xlabel('Batch Size');plt.ylabel('mAP @0.5')
plt.show()
```

### *Effect of changing the learning rate and cosine scheduler*

We experimented with constant and varying learning rates (with and without a consine scheduler). According to our findings, models trained using cosine schedulers (with initial and final learning rates opitmized for the best combination using Raytune library) performed better as compared to the others. The results are as shown below. 

```python
import pandas as pd
import matplotlib.pyplot as plt
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(y[i], i, y[i])
# mAP @0.5 AdamW s 0.005(lr0)-0.0001(lrf) 
data = {"Experiment": ["Without Scheduler", "Cosine Scheduler"],
        "mAP @0.5": [0.663, 0.664]
       }
df = pd.DataFrame(data)
```

```python
df.plot.barh(x='Experiment', y='mAP @0.5', color='brown')
plt.title('Performance of Models');addlabels(df['Experiment'],df['mAP @0.5']);plt.xlabel('Experiment');plt.xlim([0, 0.75])
plt.ylabel('mAP @0.5')
plt.show()
```

For the above experiment, we can conclude that the cosine scheduler had improved the performance as compared to without using it. 

### *Effect of increasing the number of epochs*

For the initial experiments, the number of epochs were restricted to merely 20-30 epochs inorder to save up computational resources. But what if we increased the number of epochs keeping other hyperparameters the same? According to our experiments, as we increased the number of epochs, the model improved slightly on every epoch. Models that were run for more than 50 epochs performed marginally better than those who were trained on just 20 epochs. The detailed results are as shown below. 

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {"Number of epochs": [0, 10, 20, 30, 50, 100, 150],
        "mAP @0.5": [0, 0.629,0.671, 0.736, 0.766, 0.781, 0.782]
       }
df = pd.DataFrame(data)
```

```python
plt.plot(df['Number of epochs'], df['mAP @0.5'], color='red')
plt.title('Performance of Models');plt.xlabel('Number of epochs');plt.ylabel('mAP @0.5')
plt.show()
```

The results showed that as we increase the number epochs the model's mAP increased. That is mainly due to the face that the model is gradually approaching the local minimum of the loss function. The performance of the model is going to be improved as we increase the number of epochs. However, as we increase the number of epochs, the equivalent increment in performance starts to decline to a point where the increment is almost negligible.

### *Effect of changing the model size*

YoloV8 object detection model has different ranges of sizes ranging from nano to extra large. As we increase the model size, we also increase the layers of CNN used, and training time. But that reduces the speed of inference. The results showed that as we increase the model size, the performance of the model improved (sometimes drastically, and other times by a small margin). 
The details are as shown below. 

```python
import pandas as pd
import matplotlib.pyplot as plt
def addlabels(x,y):
    time_per_epoch = ["1:17","1:29","6:19","13:18"]
    index=0
    for i in range(len(x)):
        plt.text(y[i], i+0.04, time_per_epoch[i])
        index+=1
def addlabels2(x,y):
    for i in range(len(x)):
        plt.text(y[i], i-0.15, y[i])
data = {"mAP @0.5": [0.613, 0.657, 0.776, 0.783],
#         "Time per epoch": ["1:17","1:29","6:19","13:18"] == seconds [77,89,379,798] scaled version below
        "Time per epoch":[0.096, 0.112, 0.475,1.000]
       }
df = pd.DataFrame(data, columns=["mAP @0.5", "Time per epoch"], index=["Nano (n)", "Small (s)", "Medium (m)", "Large (l)"])
```

```python
# print(df)
df.plot.barh()
plt.title('Performance of Models');plt.xlabel('Model Size');addlabels(df["mAP @0.5"], df["Time per epoch"])
addlabels2(df["Time per epoch"], df["mAP @0.5"]);plt.ylabel('mAP @0.5');plt.xlim([0, 1.3])
plt.show()
```

As we increase the model size, we are practically increasing the number of CNN layers. It takes more time, and memory to train and evaluate models who are of a bigger size. Hence, we were forced to decrease the batch size for models above m. The larger models always performed better than their smaller counterparts. 
When using larger models, there are more parameters to tune and it takes more time to tune those parameters. Additionally, they require a larger dataset. Neverthless, since we used more than 14000 images for training, we don't need to worry about that.

For models above medium, the change was more or less much lower than expected. The model took double the time when training per epoch, but it didn't show a significant change in performance. The larger the model size, the higher the inference time. Therefore, we selected medium size as it has good trade-offs between performance and inference time (as shown above).
